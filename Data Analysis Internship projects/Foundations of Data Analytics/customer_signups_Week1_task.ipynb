```python
# 3. Are older users more or less likely to opt in to marketing?
df['age'] = df['age'].str.replace('unknown', '')
df['age'] = df['age'].str.replace('thirty', '30')
df['age']
```

```python
# Dropping the empty entries from the age column 
for x in df.index:
    if df.loc[x, "age"] == '':
        df.drop(x, inplace = True)
        
df
```

```python
# Converting age column string values into numeric column
df['age'] = pd.to_numeric(df['age'])
```

```python
# Age summary
df5 = df['age']
print("Age summary \n", df5.describe())
```

```python
# Splitting the dataframe into two - one for ages below 50 and the other for ages above 50
df['age_group'] = df['age'].apply(lambda x: 'Below 50' if x < 50 else '50 and Above')

grouped = df.groupby('age_group')

# Access each group
below_50 = grouped.get_group('Below 50')
above_50 = grouped.get_group('50 and Above')
above_50
```

```python
df_marketing = above_50.groupby('marketing_opt_in')['age'].count()
print("Marketing opt-in counts by older users \n")
df_marketing
```

```python
# 4. Which plan is most commonly selected, and by which age group? 
df_plan_below50 = below_50.groupby('plan_selected')['age'].count()
print("Plans selected by users in the age category below_50: \n")
df_plan_below50
```

```python
df_plan_above50 = above_50.groupby('plan_selected')['age'].count()
print("Plans selected by users in the age category above_50: \n")
df_plan_above50
```
